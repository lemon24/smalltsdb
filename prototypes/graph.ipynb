```python
from bokeh.plotting import figure 
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.palettes import RdYlBu4
output_notebook()

import pandas as pd
```

```python
from smalltsdb import SmallTSDB
tsdb = SmallTSDB(':memory:')
tsdb.insert([
    ('one', 5, 2),
    ('two', 6, 4),
    ('one', 8, 1),
    ('two', 12, 4),
    ('one', 16, 5),
    ('one', 22, 2),
    ('one', 31, 1),
    ('one', 33, 2),
    ('two', 40, 6),
    ('one', 48, 4),
])
```

```python
df = pd.read_sql_query(
    "select * from tensecond order by path, timestamp;",
    tsdb.db,
    parse_dates=['timestamp']
)
df
```

```python
def not_good(df):
    df = df.set_index(['path', 'timestamp'])
    p = figure(x_axis_type='datetime', title='p50')
    p.line(x='timestamp', y='p50', source=ColumnDataSource(df.xs('one')), legend='one', line_width=2, line_color=RdYlBu4[0])
    p.line(x='timestamp', y='p50', source=ColumnDataSource(df.xs('two')), legend='two', line_width=2, line_color=RdYlBu4[1])
    show(p)

not_good(df)
```
